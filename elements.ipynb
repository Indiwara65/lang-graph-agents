### State           
* Current state of our applications.       
* State is simply a data structure that holds the current information or context of the entire application.

### Nodes        
* Nodes a specific functions or oprations that perform specifc task.          
* A *Node* takes an input (often the current state) and output an updated state.

### Graph        
* Graph is the structure or the map that connects the nodes and creates a excution plan for the nodes.                


### Edges        
* Edges are what connects two nodes togeather in a graph.                             
* It determines the flow of execution.

### Conditional Edge              
* Special type of edge that takes a decision on which node to execute next based on the output state of the previous node.

### Start         
* Entry point of a graph.   
* This is where the execution of the graph begin.

### End          
* Exit point of graph.        
* This where the graphs execution ends.

### Tools        
* Tools are utilities that a note can utilize to generate the next state.

### State Graph
* A state graph is like a blue print of the graph.       
* StateGraph is a class in LangGraph used to build and compile the graph structure.        
* It manages nodes, edges and the overall state.
